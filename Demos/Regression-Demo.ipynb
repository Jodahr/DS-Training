## Models for Regression

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = make_regression(n_samples=40, n_features=1, n_informative=1, noise=30, random_state=42)
```

```python
features = data[0]
target = data[1]
```

```python
df = pd.DataFrame(data[0], columns=['x'])
df['y'] = target
```

```python
sns.lmplot(x='x', y='y', data=df)
```

```python
sns.residplot(x='x', y='y', data=df)
```

```python
import matplotlib
font = {
        'size'   : 14}

matplotlib.rc('font', **font)

fig, axes = plt.subplots(2, 1, figsize=(8,8), sharex=True)
fig.subplots_adjust(hspace=0)
sns.regplot(x='x', y='y', data=df, ax=axes[0])
sns.residplot(x='x', y='y', data=df, ax=axes[1])
axes[0].set_ylabel(r"y", fontsize=18)
#axes[1].set_ylabel(r"$\mathrm{y_i} - \mathrm{\hat{y}_i}$", fontsize=18)
axes[1].set_ylabel(r"residuals", fontsize=18)
```
