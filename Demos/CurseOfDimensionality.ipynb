```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
def createSamples(dim):
    """dim is a list"""
    arr = np.random.rand(*dim)
    if len(dim) == 1:
        return pd.Series(arr)
    return pd.DataFrame(arr)
```

## 1 dimension

```python
points_1D = createSamples([100])
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure()
a = [1,2,5,6,9,11,15,17,18]
plt.hlines(1,0,1)  # Draw a horizontal line
plt.eventplot(points_1D, orientation='horizontal', colors='b')
plt.axis('off')
plt.show()
```

```python
def pointsInCircle(points):
    counts = 0
    total = len(points)
    return (points_2D.loc[:,0]**2 + points_2D.loc[:,1]**2 <= 1).sum() / total
```

```python
pointsInCircle(points_2D)
```

```python
def pointsInSphere(points):
    dim = len(points.columns)
    sum = 0
    for element in range(0,dim):
        sum+=points.loc[:,element]**2
    return (sum < 1).sum() / len(points)
```

```python
pointsInSphere(points_2D)
```
